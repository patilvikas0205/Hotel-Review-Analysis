```python
all_review=[]
for review in df["Review"]:
    msg=re.sub('[^a-zA-Z]'," ", review)
    msg.lower()
    msg1=[ps.stem(wordnet_lemmatizer.lemmatize(word,pos="n")) for word in msg.split(" ") if not word in stopwords.words("english")]
    msg1=set(msg1)
    pmsg=" ".join(msg1)
    all_review.append(pmsg)
```

```python
all_review
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
X=cv.fit_transform(all_review).toarray()
```

```python
X.shape
```

```python
y=df["Liked"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=9)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model=RandomForestClassifier()
rf_model.fit(X_train,y_train)
```

```python
rf_model.score(X_train,y_train)
```

```python
rf_model.score(X_test,y_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
nv_model=MultinomialNB()
nv_model.fit(X_train,y_train)
nv_model.score(X_test,y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
nv_model=GaussianNB()
nv_model.fit(X_train,y_train)
nv_model.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,nv_model.predict(X_test))
```
